```python
from pycolleff.hhcavity import Params, HarmonicCavity
import mathphys
import numpy as np
```

```python
hc_params = Params()
hcavity = HarmonicCavity()

hc_params.sirius_params()
hc_params.Q = 2e8
# hc_params.maxiv_params()
hcavity.params = hc_params
```

```python
hcavity.print_flat_potential()
```

```python
dt = 0.9
frf = hcavity.params.frf
c = mathphys.constants.light_speed
npts = int(1e6)
tau = np.linspace(-1/frf/2 * dt, 1/frf/2 * dt,  npts)
z = tau * c
```

```python
fig = hcavity.plot_voltages(z)
```
